```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BRCA1=pd.read_csv('/Users/joanacorreia/Desktop/Tese/Processed Tables/RH/BRCA1/RH_BRCA1_processed.csv')
```

```python
BRCA1['Gene'].unique()
```

```python
BRCA1_intergenic=BRCA1[BRCA1['Gene']== 'Intergenic control']
numeric_BRCA1_intergenic = BRCA1_intergenic.select_dtypes(include=['number'])

plt.figure(figsize=(15, 6))
sns.boxplot(data=numeric_BRCA1_intergenic)
plt.xticks(rotation=90)
plt.title("Boxplot of RH Intergenic Control Counts")
plt.show()

BRCA1_non_targeting=BRCA1[BRCA1['Gene']== 'Non-targeting control']
numeric_BRCA1_non_targeting = BRCA1_non_targeting.select_dtypes(include=['number'])

plt.figure(figsize=(15, 6))
sns.boxplot(data=numeric_BRCA1_non_targeting)
plt.xticks(rotation=90)
plt.title("Boxplot of RH Non Targeting Control Counts")
plt.show()

BRCA1_panlethal=BRCA1[BRCA1['Gene']== 'Panlethal splice donor']
numeric_BRCA1_panlethal = BRCA1_panlethal.select_dtypes(include=['number'])

plt.figure(figsize=(15, 6))
sns.boxplot(data=numeric_BRCA1_panlethal)
plt.xticks(rotation=90)
plt.title("Boxplot of RH Panlethal Control Counts")
plt.show()
```
