# merge total


```python
RH=pd.read_csv("data/4_Screen_zscores/RH/merge_RH_z.csv")
GR=pd.read_csv("data/4_Screen_zscores/GR/merge_GR_z.csv")
MC=pd.read_csv("data/4_Screen_zscores/MC/MC_zscores.csv")
```

```python
RH=RH.drop(columns='Unnamed: 0')
GR=GR.drop(columns='Unnamed: 0')
```

```python
RH.rename(columns={col: col + "_RH" if col in RH.columns[3:] else col for col in RH.columns}, inplace=True)
GR.rename(columns={col: col + "_GR" if col in GR.columns[3:] else col for col in GR.columns}, inplace=True)
MC.rename(columns={col: col + "_MC" if col in MC.columns[3:] else col for col in MC.columns}, inplace=True)
```

```python
GR.columns
```

```python
MC.columns
```

```python
#Manter a ordem original
RH["order"] = RH.index 
RH_GR = RH.merge(GR, on=["Guide", "Gene", "Editor"], how="outer", suffixes=("_RH", "_GR"))
RH_GR = RH_GR.sort_values(by=["order"]).drop(columns=["order"]) 
```

```python
RH_GR["order"]=RH_GR.index
complete_merge=RH_GR.merge(MC, on=["Guide", "Gene","Editor"],how='outer')
complete_merge = complete_merge.sort_values(by=["order"]).drop(columns=["order"]) 
```

```python
complete_merge.to_csv("data/complete_merge.csv")
```
