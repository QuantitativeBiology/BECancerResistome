# RH merge

```python
import numpy as np
import pandas as pd
from dotenv import load_dotenv
import os

load_dotenv()
base_dir = os.getenv("BASE_DIR")
```

```python
a=pd.read_csv(os.path.join(base_dir,"3_Counts/RH/RH_BRCA2_processed.csv"))
b=pd.read_csv(os.path.join(base_dir,'3_Counts/RH/RH_BRCA2_processed.csv'))
c=pd.read_csv(os.path.join(base_dir,'3_Counts/RH/RH_MCL1_processed.csv'))
d=pd.read_csv(os.path.join(base_dir,'3_Counts/RH/RH_BCL2L1_processed.csv'))
e=pd.read_csv(os.path.join(base_dir,'3_Counts/RH/RH_PARP1_processed.csv'))

```

```python
a["Gene"] = a["Gene"].fillna("Unknown")
a['Editor']=  a["Editor"].fillna("Unknown")
b["Gene"] = b["Gene"].fillna("Unknown")
b['Editor']=  b["Editor"].fillna("Unknown")
c["Gene"] = c["Gene"].fillna("Unknown")
c['Editor']=  c["Editor"].fillna("Unknown")
d["Gene"] = d["Gene"].fillna("Unknown")
d['Editor']=  d["Editor"].fillna("Unknown")
e["Gene"] = e["Gene"].fillna("Unknown")
e['Editor']=  e["Editor"].fillna("Unknown")
```

```python
merge_BRCA=pd.concat([a, b], ignore_index=True)
merge_2=pd.concat([c, d], ignore_index=True)
merged_RH=pd.concat([merge_BRCA,merge_2,e], ignore_index=True)

merged_RH.to_csv(os.path.join(base_dir,'3_Counts/RH/merge_RH.csv'))
```
