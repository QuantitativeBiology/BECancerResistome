```python
merge_BRCA=pd.concat([a, b], ignore_index=True)
merge_2=pd.concat([c, d], ignore_index=True)
final_RH=pd.concat([merge_BRCA,merge_2,e], ignore_index=True)
#final_RH.to_csv('/Users/joanacorreia/Desktop/Tese/Processed Tables/RH/merge_RH.csv')
```
