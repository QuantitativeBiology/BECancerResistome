## attempt of AUC and PRC on RH

```python
BRCA1_zscores=pd.read_csv('/Users/joanacorreia/Desktop/Tese/Processed Tables/RH/BRCA1/RH_BRCA1_Zscores.csv')
BRCA1_PRC = BRCA1_zscores[BRCA1_zscores['Gene'].isin(['Intergenic control', 'Panlethal splice donor'])]

BRCA1_PRC["True_Label"] = np.where(BRCA1_PRC["Gene"] == "Panlethal splice donor", 0, 1)
score_columns = [col for col in BRCA1_PRC.columns if col not in ["Gene","Guide","Editor","pDNA", "True_Label"]]

for col in score_columns:
    y_scores = BRCA1_PRC[col]  
    y_true = BRCA1_PRC["True_Label"] 

    precision, recall, _ = precision_recall_curve(y_true, y_scores)

    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)  

    plt.figure(figsize=(6, 6))
    plt.plot(recall, precision, marker='.', label=f"PR Curve ({col})")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title(f"Precision-Recall Curve for {col}")
    plt.legend()
    plt.show()

    plt.figure(figsize=(6, 6))
    plt.plot(fpr, tpr, marker='.', label=f'ROC Curve ({col}, AUC = {roc_auc:.2f})')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve for {col}")
    plt.legend()
    plt.show()


#if we wanted smooth
# scaler = MinMaxScaler()
# BRCA1_PRC[score_columns] = scaler.fit_transform(BRCA1_PRC[score_columns])

# for col in score_columns:
#     y_scores = BRCA1_PRC[col]  
#     y_true = BRCA1_PRC["True_Label"]  

#     precision, recall, _ = precision_recall_curve(y_true, y_scores)

#     fpr, tpr, _ = roc_curve(y_true, y_scores)
#     roc_auc = auc(fpr, tpr) 

#     # Smooth PR Curve (Moving Average)
#     precision_smooth = uniform_filter1d(precision, size=5)
#     recall_smooth = uniform_filter1d(recall, size=5)

#     plt.figure(figsize=(6, 6))
#     plt.plot(recall_smooth, precision_smooth, marker='.', label=f"Smoothed PR ({col})")
#     plt.xlabel("Recall")
#     plt.ylabel("Precision")
#     plt.title(f"Precision-Recall Curve for {col}")
#     plt.legend()
#     plt.show()

#     plt.figure(figsize=(6, 6))
#     plt.plot(fpr, tpr, marker='.', label=f'ROC Curve ({col}, AUC = {roc_auc:.2f})')
#     plt.xlabel("False Positive Rate")
#     plt.ylabel("True Positive Rate")
#     plt.title(f"ROC Curve for {col}")
#     plt.legend()
#     plt.show()

#     print(f"{col} - AUC-ROC: {roc_auc:.2f}")
```
