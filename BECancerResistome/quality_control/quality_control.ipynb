```python
MC_QC=pd.read_excel('/Users/joanacorreia/Desktop/Tese/Processed Tables/MC/41588_2024_1948_MOESM4_ESM.xlsx',sheet_name='ST1 base_editing')
MC_QC=MC_QC[['Gene','guide','sgRNA_type']]
MC_QC.rename(columns={'guide': 'Guide'}, inplace=True)
MC_QC["Gene"] = MC_QC["Gene"].fillna("Unknown")
print(MC_QC['sgRNA_type'].unique())


MC=MC.merge(MC_QC, on=["Guide","Gene"],how='left')
#Replace controls in Gene column
MC["Gene"] = MC.apply(lambda row: row["sgRNA_type"] if row["sgRNA_type"] != "exonic" else row["Gene"], axis=1)
MC = MC.drop(columns=["sgRNA_type"])
MC['Gene'].nunique()
MC
```

```python
column_name = 'P_C9_Gefit_RepB'  
threshold_value = 60000       
mean_of_rest = MC[column_name][MC[column_name] <= threshold_value].mean()
MC.loc[MC[column_name] > threshold_value, column_name] = mean_of_rest
```

```python
MC_splice_essential=MC[MC['sgRNA_type']== 'splice_essential']
numeric_MC_splice_essential = MC_splice_essential.select_dtypes(include=['number'])

plt.figure(figsize=(15, 6))
sns.boxplot(data=numeric_MC_splice_essential)
plt.xticks(rotation=90)
plt.title("Boxplot of MC Splice essential Control Counts")
plt.show()

MC_splice_non_essential=MC[MC['sgRNA_type']== 'splice_nonessential']
numeric_MC_splice_non_essential = MC_splice_non_essential.select_dtypes(include=['number'])

plt.figure(figsize=(15, 6))
sns.boxplot(data=numeric_MC_splice_non_essential)
plt.xticks(rotation=90)
plt.title("Boxplot of MC Splice non essential Control Counts")
plt.show()
```
