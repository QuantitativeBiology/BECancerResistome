### Ensembl


```python
def count_ng_pam(gene_sequence):
    pam_length = 3
    sgRNA_count = 0

    for i in range(len(gene_sequence) - pam_length):
        pam = gene_sequence[i : i + pam_length]

        if pam[1] == "G" and pam[2] == "G":
            sgRNA_count += 1

    return sgRNA_count
```

```python
def count_ng_pams_in_genes(fasta_file, gene_list):
    ng_counts = {}

    # Initialize counts for each gene in the gene list
    for gene in gene_list:
        ng_counts[gene] = 0

    # Read the FASTA file and count NG PAMs
    for record in SeqIO.parse(fasta_file, "fasta"):
        # Extract gene information from the FASTA header
        header = record.description

        gene_name = header.split("gene_symbol:")

        if len(gene_name) == 1:
            continue

        gene_name = gene_name[1].split()[0]  # Extract gene symbol

        # Check if the gene is in the specified list
        if gene_name in gene_list:
            gene_sequence = str(record.seq)
            # Count NG PAMs in the gene sequence
            count = count_ng_pam(gene_sequence)
            ng_counts[gene_name] += count

    return ng_counts
```

```python
fasta_file = "Homo_sapiens.GRCh38.cds.all.fa"
```

```python
# Count NG PAM occurrences
ng_pam_counts = count_ng_pams_in_genes(fasta_file, genes)

# Print the results
for gene, count in ng_pam_counts.items():
    print(f"{gene}: {count} NG PAM occurrences")
```

### Uniprot


```python
def calc_NG(aa_len: int):
    return aa_len * 3 * 0.5 * 0.8
```

```python
hngc = pd.read_csv("protein-coding_gene.txt", sep="\t")
hngc_uniprot = set(hngc["uniprot_ids"])
hngc_gene_symbol = hngc.groupby("uniprot_ids")["symbol"].first()
hngc
```

```python
from Bio import SeqIO


def read_fasta_to_dict(fasta_file):
    sequences = {}
    for record in SeqIO.parse(fasta_file, "fasta"):
        if "OX=9606" in record.description:
            sequences[record.id] = str(record.seq)
    return sequences


uniprot_sequences = read_fasta_to_dict("uniprot_sprot.fasta")
len(uniprot_sequences)
```

```python
gene_aa = {
    hngc_gene_symbol.loc[k.split("|")[1]]: v
    for k, v in uniprot_sequences.items()
    if k.split("|")[1] in hngc_uniprot
}
gene_aa
```

```python
genes = [
    "EGFR",
    "KRAS",
    "NRAS",
    "HRAS",
    "BRAF",
    "RAF1",
    "ERBB3",
    "ERBB2",
    "MAPK1",
    "MAPK3",
    "MAP2K1",
    "MAP2K2",
    "PIK3CA",
    "PIK3CB",
    "PIK3CD",
    "AKT1",
    "MYC",
    "SHC1",
    "GRB2",
    "GAB1",
    "ITGA3",
    "IGF1R",
    "FGFR2",
    "MTOR",
    "RICTOR",
    "NF1",
    "PTEN",
    "PKM",
    "PHLDA2",
    "PGD",
]
```

```python
lib_genes = pd.DataFrame(
    [
        dict(gene=g, aa=len(gene_aa[g]), NG=calc_NG(len(gene_aa[g])))
        for g in genes
    ]
)
lib_genes.sort_values("aa", ascending=False)
```
