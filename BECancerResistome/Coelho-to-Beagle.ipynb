```python
import pandas as pd

# Open Coelho et al. publication z-score file

# Specify the path to your Excel file
file_path = "../Publication_files/41588_2024_1948_MOESM4_ESM.xlsx"

# Load the specified tab "ST2 BE z-scores" into a DataFrame
data = pd.read_excel(file_path, sheet_name="ST2 BE z-scores")

# Keep only the specified columns
columns_to_keep = [
    "Gene",
    "editor",
    "guide",
    "L2FC_HT29_DebCet_plasmid_average_zscore",
    "L2FC_HT29_Tram_plasmid_average_zscore",
    "L2FC_HT29_Pict_plasmid_average_zscore",
    "L2FC_HT29_Control_plasmid_average_zscore",
    "L2FC_H23_Adag_plasmid_average_zscore",
    "L2FC_H23_Sotor_plasmid_average_zscore",
    "L2FC_H23_Control_plasmid_average_zscore",
    "L2FC_PC9_Control_plasmid_average_zscore",
    "L2FC_PC9_Osim_plasmid_average_zscore",
    "L2FC_PC9_Gefit_plasmid_average_zscore",
    "L2FC_MHHES1_Olap_plasmid_average_zscore",
    "L2FC_MHHES1_Nirap_plasmid_average_zscore",
    "L2FC_MHHES1_Control_plasmid_average_zscore",
]
filtered_data = data[columns_to_keep]

# Save the filtered data to a new Excel file
# output_file_path = "Filtered_z-sores.xlsx"
# filtered_data.to_excel(output_file_path, index=False)

# print(f"Filtered data saved to {output_file_path}")

filtered_data
```

```python
# Open Beagle generated libraries

# Specify the path to your file
file_path = "../Beagle output/EG_collab-guides.txt"
# Load the tab-delimited text file into a DataFrame
data_Beagle = pd.read_csv(file_path, sep="\t")

print(data_Beagle.columns.tolist())
```

```python
print(data_Beagle.columns.tolist())
```

```python
# Mering files for ABE
filtered_data_ABE = filtered_data[data["editor"].str.contains("ABE", na=False)]
data_Beagle_ABE = data_Beagle[data_Beagle["Edit Type"].str.contains("A-G", na=False)]

# Merge the datasets, preserving all rows from data_Beagle_ABE
merged_data_ABE = data_Beagle_ABE.merge(
    filtered_data_ABE, left_on="sgRNA Sequence", right_on="guide", how="left"
)

# Save the merged data to an Excel file
merged_data_ABE.to_excel("EG_paper_merged_to_Beagle_ABE.xlsx", index=False)

# Display the merged dataset
print(merged_data_ABE.columns.tolist())
```

```python
# Mering files for CBE
filtered_data_CBE = filtered_data[data["editor"].str.contains("CBE", na=False)]
data_Beagle_CBE = data_Beagle[data_Beagle["Edit Type"].str.contains("C-T", na=False)]

# Merge the datasets, preserving all rows from data_Beagle_CBE
merged_data_CBE = data_Beagle_CBE.merge(
    filtered_data_CBE, left_on="sgRNA Sequence", right_on="guide", how="left"
)

# Save the merged data to an Excel file
merged_data_CBE.to_excel("EG_paper_merged_to_Beagle_CBE.xlsx", index=False)

# Display the merged dataset
print(merged_data_CBE)
```
