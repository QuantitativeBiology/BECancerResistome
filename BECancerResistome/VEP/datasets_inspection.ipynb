## Inspect Beagle Files

### Check for column inconsistencies

```python
print("MC Beagle Columns:" , mc_beagle.columns.to_list())
print("GR Beagle Columns:" , gr_beagle.columns.to_list())
print("EG Beagle Columns:" , eg_beagle.columns.to_list())
print("RH BCL2L1_MCL1 Beagle Columns:" , rh_bcl2l1_mcl1_beagle.columns.to_list())
print("RH BRCA1_BRCA2 Beagle Columns:" , rh_brca1_brca2_beagle.columns.to_list())
print("RH PARP1 Beagle Columns:" , rh_parp1_beagle.columns.to_list())
```

```python
column_dict = {
    "MC": mc_beagle.columns,
    "GR": gr_beagle.columns,
    "EG": eg_beagle.columns,
    "RH_BCL2L1_MCL1": rh_bcl2l1_mcl1_beagle.columns,
    "RH_BRCA1_BRCA2": rh_brca1_brca2_beagle.columns,
    "RH_PARPI": rh_parp1_beagle.columns
}

# MC as reference dataset 
reference = column_dict["MC"]

# Check differences
for name, cols in column_dict.items():
    if not reference.equals(cols):
        print(f"\n{name} has different columns from MC")
        print(" Extra columns:", set(cols) - set(reference))
        print(" Missing columns:", set(reference) - set(cols))
    else:
        print(f"\n{name} columns match MC")
```

#### Standardize GR columns to match the others

NOTE: If loading the standardized GR Beagle dataset, do not run this section

```python
gr_beagle.head()
```

```python
# Separate 'Target Assembly' and 'Target Taxon' columns
gr_beagle[['Target Assembly', 'Target Taxon']] = gr_beagle['Target Assembly'].str.extract(r'(\w+)\s*\((\d+)\)')

#Rename columns names
gr_beagle.rename(columns={
    'sgRNA Target Sequence': 'sgRNA Sequence',
    'sgRNA Target Sequence Start Pos. (global)': 'sgRNA Sequence Start Pos. (global)' 
    }, inplace=True)

# Correct colmn positions
gr_beagle = gr_beagle[['Input', 'CRISPR Enzyme', 'Edit Type', 'Edit Window', 'Target Taxon', 'Target Assembly', 'Target Genome Sequence', 'Target Gene ID', 'Target Gene Symbol', 'Target Gene Strand', 'Target Transcript ID', 'Target Domain', 'sgRNA Sequence', 'sgRNA Context Sequence', 'PAM Sequence', 'sgRNA Sequence Start Pos. (global)', 'sgRNA Orientation', 'Nucleotide Edits (global)', 'Guide Edits', 'Nucleotide Edits', 'Amino Acid Edits', 'Mutation Category', 'Constraint Violations', 'Note']]
```

```python
gr_beagle.to_csv("data/1_Beagle_design_files/GR/GR-guides-standardized.txt", sep="\t", index=False)
```
