### Filter files by editor (ABE & CBE)

```python
#Filter files for ABE
mc_zscore_abe = mc_zscore[mc_zscore["Editor"].str.contains("ABE", na=False)]
mc_beagle_abe = mc_beagle[mc_beagle['Edit Type'].str.contains("A-G", na=False)]
mc_beagle_utr_abe = mc_beagle_utr[mc_beagle_utr['Edit Type'].str.contains("A-G", na=False)]

#Filter files for CBE
mc_zscore_cbe = mc_zscore[mc_zscore["Editor"].str.contains("CBE", na=False)]
mc_beagle_cbe = mc_beagle[mc_beagle['Edit Type'].str.contains("C-T", na=False)]
mc_beagle_utr_cbe = mc_beagle_utr[mc_beagle_utr['Edit Type'].str.contains("C-T", na=False)]
```
